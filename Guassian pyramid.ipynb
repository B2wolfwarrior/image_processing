```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread(r"C:\Users\BhavyeBhasin\Downloads\SS_007.png")
img = cv2.resize(img, (700, 700))

# Gaussian Pyramid (using cv2.pyrDown and cv2.pyrUp)
pd1 = cv2.pyrDown(img)
pd2 = cv2.pyrDown(pd1)
pu1 = cv2.pyrUp(pd2)

# Display images using Matplotlib
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(pd1, cv2.COLOR_BGR2RGB))
plt.title("PyrDown 1")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(pd2, cv2.COLOR_BGR2RGB))
plt.title("PyrDown 2")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(pu1, cv2.COLOR_BGR2RGB))
plt.title("PyrUp 1")
plt.axis("off")

plt.tight_layout()
plt.show()
"""
Gaussian Pyramid Operations:
-cv2.pyrDown: This function reduces the size of the image by a factor of two using Gaussian filtering. 
It’s applied twice, first to the original image (pd1) and then to the result of the first operation (pd2).
-cv2.pyrUp: This function increases the size of the image by a factor of two, also using Gaussian filtering. 
It’s applied to the result of the second pyrDown operation (pu1).
"""
```
