```python
"""Image segmentation using grabcut algorithm"""
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load the car image
img = cv2.imread(r"C:\Users\BhavyeBhasin\Downloads\SS_007.png")
img = cv2.resize(img, (800, 800))
mask = np.zeros(img.shape[:2], np.uint8)

bgdModel = np.zeros((1, 65), np.float64) * 255
fgdModel = np.zeros((1, 65), np.float64) * 255

# Define the rectangular region of interest
rect = (134, 150, 660, 730)

# Apply GrabCut algorithm
cv2.grabCut(img, mask, rect, bgdModel, fgdModel, 5, cv2.GC_INIT_WITH_RECT)

# Create a binary mask (0 for background, 1 for foreground)
mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')

# Apply the mask to the original image
img_result = img * mask2[:, :, np.newaxis]

# Display the result
plt.figure(figsize=(8, 8))
plt.imshow(cv2.cvtColor(img_result, cv2.COLOR_BGR2RGB))
plt.title("GrabCut Result")
plt.axis("off")
plt.show()
""" ROI is segmented out from the background, demonstrating an application of the 
GrabCut algorithm for image segmentation tasks."""
```
