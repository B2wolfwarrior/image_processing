```python
"""multiple roi"""
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread(r"C:\Users\BhavyeBhasin\Downloads\SS_007.png")
img = cv2.resize(img,(800,800))

# Convert the image from BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Now extract area of interest from an image
# Now pass like [y1:y2,x1:x2]
roi = img_rgb[50:205,320:440]

# Putting roi into any pixel values
img_rgb[50:205,431:551] = roi   # Actual 441:561
img_rgb[50:205,552:672] = roi 
img_rgb[50:205,200:320] = roi
img_rgb[50:205,80:200] = roi

# Changing y===
img_rgb[400:555,60:180] = roi

# Now try to use one image data into another image
img1 = cv2.imread(r"C:\Users\BhavyeBhasin\Downloads\SS_007.png")
img1 = cv2.resize(img1,(900,600))

# Convert the image from BGR to RGB
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img1_rgb[1:156,560:680] = roi

# Display images using matplotlib
plt.figure(figsize=(10,7))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(img1_rgb)
plt.title('result Image')
plt.show()
```

```python
"""region of interest"""
import cv2
import matplotlib.pyplot as plt

# Load an image using OpenCV
img = cv2.imread(r"C:\Users\BhavyeBhasin\Downloads\wolfss.png")

# Convert the image from BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Define the Region of Interest (ROI)
# Here we are defining ROI as a rectangle for simplicity
# You can define it according to your requirements
roi = img_rgb[17:346, 255:570]
# Display the original image and ROI using matplotlib
plt.figure(figsize=(10,7))
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(roi)
plt.title('Region of Interest')
plt.show()
```
