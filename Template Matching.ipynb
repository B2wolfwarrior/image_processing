```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the original image
img = cv2.imread(r"C:\Users\BhavyeBhasin\Downloads\wolfss.png")
grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Load the template image
template = cv2.imread(r"C:\Users\BhavyeBhasin\Downloads\wolfss.png", 0)
w, h = template.shape[::-1]

# Match the template using TM_CCORR_NORMED method
res = cv2.matchTemplate(grey_img, template, cv2.TM_CCORR_NORMED)
threshold = 0.999
loc = np.where(res >= threshold)

# Draw rectangles around the detected regions
count = 0
for i in zip(*loc[::-1]):
    cv2.rectangle(img, i, (i[0] + w, i[1] + h), (0, 0, 255), 2)
    count += 1

# Resize images for display
img = cv2.resize(img, (800, 600))
res = cv2.resize(res, (800, 600))

# Display the results
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Detected Regions")
plt.axis("off")

plt.subplot(122)
plt.imshow(res, cmap="gray")
plt.title("Matching Result")
plt.axis("off")

plt.tight_layout()
plt.show()

print("Number of detected regions:", count)
```
