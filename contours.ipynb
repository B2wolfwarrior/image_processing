```python
"""contour Detection"""
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread(r"C:\Users\BhavyeBhasin\Downloads\SS_007.png")
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(img1, 127, 255, 0)

# Find contours
cnts, hier = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Print the number of contours and hierarchy
print("Number of contours:", len(cnts))
print("Hierarchy:\n", hier)

# Draw the contours
contour_img = img.copy()
cv2.drawContours(contour_img, cnts, -1, (176, 10, 15), 4)

# Display images using Matplotlib
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(contour_img, cv2.COLOR_BGR2RGB))
plt.title("Contours")
plt.axis("off")

plt.tight_layout()
plt.show()
"""
-The contours are found using the cv2.RETR_TREE retrieval mode, 
which retrieves all of the contours and reconstructs the full 
hierarchy of nested contours.
-The output of this code will be a visualization of the original image and the 
same image with contours highlighted, providing insight into the shapes and 
structures within the image.
"""
```
