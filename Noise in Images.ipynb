```python
""" Salt and pepper Noise"""
import numpy as np
import cv2
import matplotlib.pyplot as plt

def add_salt_and_pepper_noise(image, salt_prob, pepper_prob):
    noisy_image = np.copy(image)
    # Add salt noise
    salt_mask = np.random.rand(*image.shape) < salt_prob
    noisy_image[salt_mask] = 255
    # Add pepper noise
    pepper_mask = np.random.rand(*image.shape) < pepper_prob
    noisy_image[pepper_mask] = 0
    return noisy_image
# Load an example image
d_path = r"C:\Users\BhavyeBhasin\Downloads\SS_007.png"
original_image = cv2.imread(d_path, cv2.IMREAD_GRAYSCALE)
# Define the probabilities for salt and pepper noise
salt_probability = 0.01
pepper_probability = 0.01
# Add salt-and-pepper noise to the image
noisy_image = add_salt_and_pepper_noise(original_image, salt_probability, pepper_probability)
# Display the original and noisy images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB))
plt.title('Noisy Image with Salt-and-Pepper Noise')
plt.axis('off')
plt.show()
"""
Salt and pepper noise, also known as impulse noise, is a type of noise
that appears as random white and black pixels in an image. It is caused
by sharp and sudden disturbances in the image signa

Contrary to the subtlety of Gaussian noise, salt-and-pepper noise 
manifests as randomly occurring white and black pixels scattered 
throughout an image. This type of noise is common in images acquired 
in low-light conditions or transmitted over noisy channels. Median filtering,
a non-linear filtering technique, is commonly employed to suppress salt-and-pepper
noise, effectively removing isolated outliers without blurring the overall image.
"""
```

```python
"""Gaussian Noise"""
import numpy as np
import cv2
import matplotlib.pyplot as plt
def add_gaussian_noise(image, mean=0, sigma=25):
    row, col, ch = image.shape
    gauss = np.random.normal(mean, sigma, (row, col, ch))
    noisy_image = image + gauss.astype('uint8')
    return np.clip(noisy_image, 0, 255)
# Load an example image
d_path = r"C:\Users\BhavyeBhasin\Downloads\SS_007.png"
original_image = cv2.imread(d_path)
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
# Add Gaussian noise to the image
noisy_image = add_gaussian_noise(original_image)
# Display the original and noisy images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(original_image)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(noisy_image)
plt.title('Noisy Image with Gaussian Noise')
plt.axis('off')
plt.show()
"""
Imagine a gentle sprinkle of random fluctuations across an image — that’s
Gaussian noise. It is characterized by pixel values varying according to a 
Gaussian distribution. This type of noise is prevalent in real-world images 
due to factors like sensor imperfections, transmission errors, or electronic 
interference. Gaussian noise is often addressed using smoothing filters like 
Gaussian filters or bilateral filters, which help in reducing the impact of 
these random fluctuations.

In this example:

    -The add_gaussian_noise function takes an image, mean, and sigma as parameters.
     Mean is the mean value of the Gaussian distribution, and sigma is the standard 
     deviation.
    -Random values from a normal distribution with the specified mean and sigma are 
     generated using np.random.normal.
    -These random values are added to the pixel values of the original image to 
     introduce Gaussian noise.
    -The np.clip function is used to ensure that pixel values remain within the 
     valid range of 0 to 255.
Key Points:

   - Gaussian noise is a common type of random noise that follows a normal distribution, also known as the bell curve.
   - Sigma (σ) is the standard deviation of the normal distribution, which controls the spread or “spreadiness” of the noise.
   - A higher sigma means the noise values are more spread out, resulting in a noisier image with more visible speckles.
   - A lower sigma means the noise values are more tightly clustered around the mean, resulting in a less noisy image with smoother textures.    
"""
```
